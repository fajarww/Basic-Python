# 3. Roll dice function



```python
def roll_dice_function(dice_number, dice_type):
    '''This function generates random number from selected
    number domain. The domain is depends on the type of dice.
    The dice number will decide how many number should be
    generated. This function must need two argument to works:
    dice_number and dice_type. The return of this function
    will be a tuple named dice_out.
    '''
    import random
    dice_out=()
    if dice_type == 6:
        dice=(1,2,3,4,5,6,)
        dice_out = random.choices(dice, k=dice_number)
        dice_out = tuple(dice_out)
        return dice_out
    elif dice_type == 8:
        dice=(2,3,4,8,)
        dice_out = random.choices(dice, weights=[2,2,1,1], k=dice_number)
        dice_out = tuple(dice_out)
        return dice_out
    else:
        dice=(1,5,9,)
        dice_out = random.choices(dice, weights=[4,1,1], k=dice_number)
        dice_out = tuple(dice_out)
        return dice_out
```

# 4. Check win/lose function


```python
def check_win_lose_function(dice_result_list = []):
    '''This function calculates the median and the average
    result of the player's roll.
    It needs one argument: dice_result_list, as an
    input, if the median & average of dice result list
    is equal, then it will return 'win' string, otherwise
    it return 'lose' string. Alongside the win/lose string,
    this function also return the median and average value
    to display for player.
    '''
    merged_result_list = []
    length_list = 0
    median_result = 0
    average_result = 0
    win_lose_string = ''
    for result in dice_result_list:
        result = list(result)
        for member in result:
            merged_result_list.append(member)
    merged_result_list.sort()
    length_list = len(merged_result_list)
    if length_list%2 == 0:
        median_result= (merged_result_list[int((length_list/2)-1)] + merged_result_list[int(length_list/2)])/2
        median_result = int(median_result)
    else:
        median_result = merged_result_list[int((length_list)//2)]
        median_result = int(median_result)
    for result in merged_result_list:
        average_result = average_result + result
    average_result = average_result / length_list
    average_result = int(average_result)
    if median_result == average_result:
        win_lose_string = 'win'
    else:
        win_lose_string = 'lose'
    return win_lose_string, median_result, average_result
```
